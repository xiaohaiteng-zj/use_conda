```python
import numpy as np
import pandas as pd
```

```python
stock = np.random.randint(40, 100, (100, 7))
```

```python
stock
```

```python
cols = ["语文", "数学", "英语", "化学", "物理", "生物", "政治"]
stock_df = pd.DataFrame(stock, columns=cols)
```

```python
stock_df
```

```python
data = stock_df.drop(["政治"], axis=1)
```

```python
data
```

# 索引操作

```python
data["数学"][1]
```

```python
data.loc[1, "数学"]
```

```python
data.iloc[1, 1]
```

```python
data.loc[:4, "语文":"化学"]
```

```python
data.iloc[:5, :3]
```

# 赋值操作

```python
data["物理"] = 70
```

```python
data
```

```python
data.物理 = 80
```

```python
data
```

```python
data.loc[:3, ["物理", "生物"]] = 100
```

```python
data
```

```python
data.sort_values(by="语文", ascending=False)
```

```python
df2 = data.sort_values(by=["语文", "数学"], ascending=False)
```

```python
df2
```
