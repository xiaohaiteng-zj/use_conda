# 实现基础绘图功能

## 绘制基本图像

```python
import matplotlib.pyplot as plt
import random
```

```python
#  1.准备数据，x,y轴坐标
x = range(60)
y_sh = [random.uniform(15, 18) for i in range(60)]

#  2.创建画布
plt.figure(figsize=(20, 8), dpi=100)

#  3.绘制图像
plt.plot(x, y_sh)

#  4.图像显示
plt.show()
```
