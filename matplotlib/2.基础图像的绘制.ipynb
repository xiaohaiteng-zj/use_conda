# 实现基础绘图功能

## 绘制基本图像

```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 设置显示中文字体
mpl.rcParams['font.sans-serif'] = ['SimHei']
# 设置正确显示符号
mpl.rcParams['axes.unicode_minus'] = False
```

```python
#  1.准备数据，x,y轴坐标
x = range(60)
y_sh = [random.uniform(15, 18) for i in range(60)]

#  2.创建画布
plt.figure(figsize=(20, 8), dpi=100)

#  3.绘制图像
plt.plot(x, y_sh)

#  4.图像显示
plt.show()
```

## 实现一些其他功能

```python
#  1.准备数据，x,y轴坐标
x = range(60)
y_sh = [random.uniform(15, 18) for i in range(60)]

#  2.创建画布
plt.figure(figsize=(20, 8), dpi=100)

#  3.绘制图像
plt.plot(x, y_sh)

#  3.1 添加x,y轴刻度
#  设置x,y轴刻度
x_ticks_label = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

#  修改x,y轴刻度显示
plt.xticks(x[::5], x_ticks_label[::5])   # 坐标刻度不可以直接通过字符串进行修改
plt.yticks(y_ticks[::5])

#  3.2 添加网格显示
plt.grid(True, linestyle="--", alpha=0.5)

#  3.3 添加描述信息
plt.xlabel("时间", fontsize=15)
plt.ylabel("温度", fontsize=15)
plt.title("中午11点-12点某城市温度变化图", fontsize=20)

#  3.4 图像保存
plt.savefig("./media/test.png")  # plt.show()会释放figure资源，因此保存图片要放在show()之前


#  4.图像显示
plt.show()
```
